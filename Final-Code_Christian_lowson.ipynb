```python
#points for skill players
def calculate_points(row):
    points = 0
    points += row['RYds'] // 10  # 1 point for every 10 rushing yards
    points += row['RTD'] * 6     # 6 points for every rushing touchdown
    points += row['Rec']         # 1 point for every reception
    points += row['WYds'] // 10  # 1 point for every 10 receiving yards
    points += row['WTD'] * 6     # 6 points for every receiving touchdown
    points -= row['Fmb'] * 2     # -2 points for every fumble
    points += row['2PS'] * 2     # 2 points for every two-point conversion successful (pass)
    points += row['2PT'] * 2     # 2 points for every two-point conversion successful (rush)
    return points

offense_20['FP/Szn'] = offense_20.apply(lambda row: calculate_points(row) if row['FantPos'] in ['RB', 'WR', 'TE'] else 0, axis=1)
offense_20
```

```python
#additional scoring for QB's
def calculate_points(row):
    points = 0
    if row['FantPos'] in ['QB', 'RB', 'WR', 'TE']:
        points += row['RYds'] // 10   # 1 point for every 10 rushing yards
        points += row['RTD'] * 6      # 6 points for every rushing touchdown
        points += row['Rec']          # 1 point for every reception
        points += row['WYds'] // 10   # 1 point for every 10 receiving yards
        points += row['WTD'] * 6      # 6 points for every receiving touchdown
        points -= row['Fmb'] * 2      # -2 points for every fumble
        points += row['2PS'] * 2      # 2 points for every two-point conversion successful (pass)
        points += row['2PT'] * 2      # 2 points for every two-point conversion successful (rush)
        points += row['Yds'] // 25    # 1 point for every 25 yards
        points += row['TD'] * 4       # 4 points for every touchdown
        points -= row['Int'] * 2      # -2 points for every interception
    return points


offense_20['FP/Szn'] = offense_20.apply(lambda row: calculate_points(row), axis=1)
offense_20
```

```python
offense_21['FP/Szn'] = offense_21.apply(lambda row: calculate_points(row), axis=1)

offense_22['FP/Szn'] = offense_21.apply(lambda row: calculate_points(row), axis=1)

offense_23['FP/Szn'] = offense_21.apply(lambda row: calculate_points(row), axis=1)
```

## I performed 10 unique 10-man league drafts to see what averages each player was at 

```python
all1 = pd.read_csv('Offense_20 - Fixed.csv')
all1
```

```python
all = all1.drop(all1.index[2::3])

all.reset_index(drop=True, inplace=True)

all
```

```python
#Organizing Data 
combined_rows = []
for i in range(0, len(all), 2):
    combined_rows.append(all.iloc[i:i+2].agg('-'.join))

position = pd.DataFrame(combined_rows)

position
```

```python
#Building a dictionary to determine what position a player was drafted at
name_dict = {}

# Iterate over each column of the DataFrame
for col in position.columns:
    for row in position[col]:
        parts = row.split('-')
        name = parts[-1].strip()
        number = float(parts[0])
        # If the name is not already in the dictionary, add it with an empty list
        if name not in name_dict:
            name_dict[name] = []
        name_dict[name].append(number)

name_dict
```

```python
#Test
name_dict['A. Brown']
```

```python
#Accidentaly broke up this code from the next
for col in position.columns:
    
    for row in position[col]:
        parts = row.split('-')
        name = parts[-1].strip()
        number_str = parts[0]  
        if name not in name_dict:
            name_dict[name] = []
        name_dict[name].append(number_str) 
```

```python
# Dictionary to store the average for each name
average_dict = {}

for name, numbers_str in name_dict.items():
    # Convert numbers from strings to floats
    numbers = [float(num_str) for num_str in numbers_str]
    average = sum(numbers) / len(numbers)
    average_dict[name] = average

for name, average in average_dict.items():
    print(f"Average for {name}: {average}")
```

```python
import pandas as pd

average_data = []

for name, numbers_str in name_dict.items():
    numbers = [float(num_str) for num_str in numbers_str]
    average = sum(numbers) / len(numbers)
    average_data.append({'Name': name, 'Average': average})

average_draft = pd.DataFrame(average_data)

average_draft.sort_values(by='Average',ascending=True)
```

```python
average_draft.sort_values(by='Average',ascending=True).reset_index(drop=True)
```
